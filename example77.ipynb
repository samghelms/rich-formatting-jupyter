```python
from IPython.display import (display_javascript,
                                         display_html)
from uuid import uuid1
class katex:
    def __init__(self, string):
        self._start = lambda uid: "<div id='"+uid+"'/>\n<script>"
        self._vars = "var parent, newChild;"
        self._parent = lambda uid: "parent = document.getElementById('{_id}');".format(_id=uid)
        self._body = ["newChild = document.createElement('span');",
                      "window.katex.render('"+string+"', newChild);",
                      "parent.appendChild(newChild);",
                       "console.log('test');"]
        self._body = "\n".join(self._body)
        self._end = "</script>"

    def _repr_html_(self):
        self._id = str(uuid1())
        rep = "\n".join([self._start(self._id), self._vars, self._parent(self._id), self._body, self._end])
        return rep

```

```python
from IPython.display import (display_javascript,
                                         display_html)
from uuid import uuid1
class katex:
    def __init__(self, s):
        self._s = s
        self._ids = []
        self.parent = lambda unique_id: 'var parent = document.getElementById("{unique_id}");'.format(unique_id=unique_id)
        self._body = lambda math: ["var newChild = document.createElement('span');",
                                "window.katex.render('{math}', newChild);".format(math=math),
                                "parent.appendChild(newChild);"]

    def _repr_html_(self):
        unique_id = uuid1()
        anchor = '<div id="{unique_id}"/>'.format(unique_id=unique_id)
        self._ids.append(unique_id)
        return anchor

    def _repr_javascript_(self):
        js = []
        while len(self._ids) > 0:
            i = self._ids.pop(0)
            js.append("\n".join([self.parent(i)]+self._body(self._s)))
        return "\n".join(js)
    
    def _ipython_display_(self):
        display_html(self._repr_html_(), raw=True)
        display_javascript(self._repr_javascript_(), raw=True)
```
